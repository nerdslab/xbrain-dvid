```python
import ndio.remote.neurodata as neurodata
import ndio.ramon as ramon

import numpy as np
from time import time
nd = neurodata()
```

```python
# Read data - will replace with built in ndio function soon

file_location = '/Users/graywr1/Downloads/S4_subblock/'

import glob
from PIL import Image

files = glob.glob(file_location + '/*.tif')
files.sort()
data = []
c = 0
for f in files:
    data.append(np.asarray(Image.open(f)))
    c += 1

data = np.asarray(data)
data = np.rollaxis(data,1)
data = np.rollaxis(data,2)
np.shape(data)
```

```python
data_norm = 255*(data - np.min(data))/(np.max(data)-np.min(data))
data_norm = np.uint8(data_norm)
# plot result
import ndparse as ndp
%matplotlib inline
ndp.plot(data_norm)
```
