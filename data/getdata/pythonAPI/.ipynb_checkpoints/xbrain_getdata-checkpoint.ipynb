### Import ndio (neurodata's python API)

```python
import ndio.remote.neurodata as neurodata
nd = neurodata('suppress_warnings=True')
```

### Specify range of data to download 

```python
im_token = 'S1_proj4'
anno_token = 'xbrain_rfr1'
x = [610,2010]#[610, 2010]
y = [1,2480]#[1, 2480]
z = [1300,1400]#[390, 2014]
```

### Download Image Data

```python
im = nd.get_cutout(im_token,'image', x[0],x[1],y[0],y[1],z[0],z[1],resolution = 0)

# uncomment any of the 4 lines at the bottom of this cell to download our:
    # final segmentations/cell detections (cellseg, vesselseg) 
    # probability maps (cellprob, vesselprob)
    
#cellseg = nd.get_cutout(anno_token,'cell_seg', x[0],x[1],y[0],y[1],z[0],z[1],resolution = 0)
#vesselseg = nd.get_cutout(anno_token,'vessel_seg', x[0],x[1],y[0],y[1],z[0],z[1],resolution = 0)
#cellprob = nd.get_cutout(anno_token,'cell_prob', x[0],x[1],y[0],y[1],z[0],z[1],resolution = 0)
#vesselprob = nd.get_cutout(anno_token,'vessel_prob', x[0],x[1],y[0],y[1],z[0],z[1],resolution = 0)
```

### Plot image data

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np

imgplot = plt.imshow(im[0:1400,0:2400,3],cmap="hot")
```

### Download vessel probabilities and segmentations

```python
vesselprob = nd.get_cutout(anno_token,'vessel_prob', x[0],x[1],y[0],y[1],z[0],z[1],resolution = 0)
vesselseg = nd.get_cutout(anno_token,'vessel_seg', x[0],x[1],y[0],y[1],z[0],z[1],resolution = 0)
```

### Plot vessel probability map + segmentation

```python
imgplot = plt.imshow(vesselprob[0:1400,0:2400,3],cmap="hot")
```

```python
imgplot = plt.imshow(vesselseg[0:1400,0:2400,3],cmap="hot")
```
